```python
import csv
import collections

with open('data.csv') as f:
    reader = csv.DictReader(f, ('rank', 'year', 'name', 'artist', 'genre', 'subgenre'))
    albums = []
    for row in reader:
        albums.append(row)
    albums.pop(0)
    
print (albums[0])
```

```python
# function to find by name

def findbyname(name, list_name):
    for entry in list_name:
        if name == entry['name']:
            return entry
    return None

print (findbyname('Hey Jude', songs))
```

```python
# function to find by rank

def findbyrank(num, list_name):
    for entry in list_name:
        if str(num) == entry['rank']:
            return entry
    return None

print (findbyrank(12, albums))
```

```python
# function to find by year

def findbyyear(year, list_name):
    entries = []
    for entry in list_name:
        if str(year) == entry['year']:
            entries.append(entry)
    return entries

print (findbyyear(1967, albums))
```

```python
# function to find by years

def findbyyears(start, end, list_name):
    final_list = []
    for entry in list_name:
        #print (entry['rank'])
        if int(entry['year']) >= start and int(entry['year']) <= end:
            final_list.append(entry)
    return final_list
    
print (findbyyears(1990, 1991, albums))
```

```python
# function to find by range of ranks

def findbyranks(start, end, list_name):
    final_list = []
    for entry in list_name:
        if int(entry['rank']) >= start and int(entry['rank']) <= end:
            final_list.append(entry)
    return final_list

print (findbyranks(1, 10, albums))
```

```python
# function for all titles

def all_titles(list_name):
    final_list = []
    for entry in list_name:
        final_list.append(entry['name'])
    return set(final_list)

print (all_titles(albums))
```

```python
# function for all artists

def all_artists(list_name):
    final_list = []
    for entry in list_name:
        final_list.append(entry['artist'])
    return set(final_list)

print (all_artists(songs))
```

```python
# artist with the most albums

def artist_most_albums(list_name):
    unique_artists = all_artists(list_name)
    total_artists = [entry['artist'] for entry in list_name]
    max = 0
    max_artist_name = ""

    for artist in unique_artists:
        artist_count = total_artists.count(artist)
        if artist_count > max:
            max = artist_count
            max_artist_name = artist
    return max_artist_name


print (artist_most_albums(songs))
```

```python
# most popular word in title

import string

def most_popular_word(list_name):
    unique_titles = all_titles(list_name)
    #total_artists = [album['artist'] for album in albums]
    max = 0
    max_word = ""
    cleaned_all_words = []
    all_words = []
    
    for title in unique_titles:
        all_words = (title.split(" "))
        for word in all_words:
            table = word.maketrans("","", string.punctuation)
            cleaned_all_words.append(word.translate(table))
            
    unique_words = list(set(cleaned_all_words))
    
    for word in unique_words:
        word_count = cleaned_all_words.count(word)
        if word_count > max:
            max = word_count
            max_word = word
            
    #return (unique_words)
    return max_word
        
print (most_popular_word(songs))
```

```python
# create dictionary of Top 500 songs

text_file = open('top-500-songs.txt', 'r')
    # read each line of the text file
    # here is where you can print out the lines to your terminal and get an idea 
    # for how you might think about re-formatting the data
lines = text_file.readlines()

#key values list
key_values = ["rank", "name", "artist", "year"]


songs = []

for line in lines:
    song_dict = {}
    line = line.split('\t')
    for key, value in zip(key_values, line):
        song_dict[key] = value.replace("\n", "")
    songs.append(song_dict)

print (songs)
```

```python
import json

file = open('track_data.json', 'r')
json_data = json.load(file)

for album in json_data:
    cleaned_track_songs = []
    track_songs = album['tracks']
    for track_song in track_songs:
        hyphen = track_song.rfind("-", 1)
        bracket = track_song.rfind("[", 1)
        break_indexes = [hyphen, bracket]
        if hyphen < 0 and bracket < 0:
            cleaned_track_songs.append(track_song)
        else:
            breaker = min(i for i in break_indexes if i > 0)
            track_song = track_song[:breaker]
            track_song = track_song.strip()
            cleaned_track_songs.append(track_song)
        
    album['tracks'] = list(set(cleaned_track_songs))
    
print(json_data)
```
